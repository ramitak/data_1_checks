```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# importing data from csv file
df = pd.read_csv('assets/movies_data.csv')
df.head()
```

```python
# Displaying top 10 rank movies 
top_10 = df.head(10)
top_10.head()
```

```python

#Creating variables to define the table columns name
def_certificate = top_10['certificate']
def_year = top_10['year']
```

```python
# Plotting datas 
plt.scatter(def_certificate, def_year)
plt.title('Top 250 Movies')
plt.xlabel('certificate')
plt.ylabel('Year')
plt.show()
```

```python

#Creating variables to define the table columns name
def_rank = top_10['rank']
def_year = top_10['year']
```

```python
# Plotting datas 
plt.scatter(def_rank, def_year)
plt.title('Top 250 Movies')
plt.xlabel('rank')
plt.ylabel('Year')
plt.show()
```
